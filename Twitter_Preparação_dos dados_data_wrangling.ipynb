```
twitter_dt.drop(['User_mentions', 'Hashtags', 'user_location', 'geo_coordinates'], axis=1, inplace=True) 
```

```
def find_mentioned(tweet):
    'This function will extract the twitter handles of people mentioned in the tweet'
    return re.findall('(?<!RT\s)(@[A-Za-z]+[A-Za-z0-9-_]+[\w+]*)', tweet)
def find_hashtags(tweet):
    '''This function will extract hashtags'''
    return re.findall('(#[A-Za-z]+[A-Za-z0-9-_]+[\w+]*)', tweet)
# make new columns for mentioned usernames and hashtags
twitter_dt['User_mentions'] = twitter_dt.full_text.apply(find_mentioned).astype(str).str.lower().str.replace("'", "").apply(unidecode.unidecode)
twitter_dt['Hashtags'] = twitter_dt.full_text.apply(find_hashtags).astype(str).str.lower().str.replace("'", "").apply(unidecode.unidecode)
```

```
twitter_dt["Media"].fillna(0, inplace = True)
twitter_dt.loc[twitter_dt['Media'] != 0, 'Media'] = "Presente"
twitter_dt.loc[twitter_dt['Media'] == 0, 'Media'] = "Ausente"
```

```
for col in twitter_dt.columns[twitter_dt.dtypes == 'bool']:
    twitter_dt[col] = twitter_dt[col].map({True: 'Sim', False: 'Não'})
```

```
twitter_dt.columns = ['Id', 'Tempo', 'Retweets', 'Curtidas', 'Nome_do_usuário', 'Usuário_verificado', 'Seguidores', 'Mídia', 'Tweet', 'Menções_de_usuários', 'Hashtags']
```
